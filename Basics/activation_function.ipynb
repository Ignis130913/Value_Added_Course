```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,5,50)
z=np.tanh(x)
```

```python
#softmax
plt.plot(x,z)
plt.grid()
plt.show()
```

```python
#relu
z=[max(0,i) for i in x]
plt.plot(x,z)
plt.grid()
plt.show()
```

```python
#sigmoid
z=1/(1+np.exp(-x))
plt.plot(x,z)
plt.grid()
plt.show()
```
